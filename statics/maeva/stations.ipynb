```python
import json
import os
```

```python
all_files = os.listdir()    
files = list(filter(lambda f: f.endswith('.json'), all_files))
dfs = []

for file in files:
    content = json.load(open(file))
    dfs += content
    
```

```python
len(dfs)
```

```python
base_urls = []
```

```python
for station_key in dfs:
    base_urls.append({'url':f"https://www.maeva.com/fr-fr/searchlist.php?map=1&acces_direct=1&station_cle={station_key}&etendre_min=30&trier_par=zerank", 'nb_page':1})
```

```python
base_urls
```

```python
with open('stations.json', 'w') as openfile:
    openfile.write(json.dumps(base_urls))
```
