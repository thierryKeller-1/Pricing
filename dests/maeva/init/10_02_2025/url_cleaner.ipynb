```python
import pandas as pd

def get_saturdays(start_date, end_date):
    saturdays = pd.date_range(start=start_date, end=end_date, freq='W-SAT')
    return saturdays
# Example usage
start_date = '2025-03-01'
end_date = '2025-03-31'
saturdays = get_saturdays(start_date, end_date)
print(saturdays)
```

```python
def add_date_params(urls, start_date, end_date):
    updated_urls = []
    for url in urls:
        if '.php' in url:
            updated_url = f"{url}&date_debut={start_date}&date_fin={end_date}"
        else:
            residence_cle = url.split('_')[-1].split('.')[0]
            updated_url = f"{url}?date_debut={start_date}&date_fin={end_date}&residence_cle={residence_cle}"
        updated_urls.append(updated_url)
    return updated_urls

# Example usage
urls = [
    "https://www.maeva.com/fr-fr/appartement-risco_819661.html",
    "https://www.maeva.com/fr-fr/pages/fiche.php?id=1153824",
    "https://www.maeva.com/fr-fr/residence-pierre-vacances-premium-arc-1950-le-village_14581.html"
]
start_date = '2025-03-01'
end_date = '2025-03-06'
updated_urls = add_date_params(urls, start_date, end_date)
print(updated_urls)
```

```python
from datetime import datetime, timedelta
import pandas as pd

def normalize_urls(urls, start_at:str, end_at:str) -> list:
    
    dates = pd.date_range(start=start_at, end=end_at, freq='C').strftime('%Y-%m-%d').to_list()
    updated_urls = []
    for date in dates:
        start_date = date 
        end_date = (datetime.strptime(date, '%Y-%m-%d') + timedelta(days=6)).strftime('%Y-%m-%d')
        for url in urls:
            if '.php' in url:
                updated_url = f"{url}&date_debut={start_date}&date_fin={end_date}"
            else:
                residence_cle = url.split('_')[-1].split('.')[0]
                updated_url = f"{url}?date_debut={start_date}&date_fin={end_date}&residence_cle={residence_cle}"
            updated_urls.append(updated_url)
    return updated_urls
```

```python
urls = [
    "https://www.maeva.com/fr-fr/appartement-risco_819661.html",
    "https://www.maeva.com/fr-fr/pages/fiche.php?id=1153824",
    "https://www.maeva.com/fr-fr/residence-pierre-vacances-premium-arc-1950-le-village_14581.html"
]

normalize_urls(urls, '22/02/2025', '30/03/2025')
```
